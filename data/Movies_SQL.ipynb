```python
from sqlalchemy import create_engine

from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
from sqlalchemy import Column, Integer, String, Float
```

```python
from sqlalchemy import create_engine
database_path = "movie_db.sqlite"
engine = create_engine(f"sqlite:///{database_path}")
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
# create your shark class
class Movies(Base):
    __tablename__ = 'sqldata'
    movie_title = Column(String, primary_key=True)
    director_name = Column(String)
    duration = Column(Integer)
    genres = Column(String)
    actors = Column(String)
    plot_keywords= Column(String)
    language = Column(String)
    country= Column(String)
    title_year = Column(Integer)
    imdb_score = Column(Integer)
```

```python
# print all locations of shark attacks
attacks = session.query(Movies) 
```

```python
for attack in attacks:
    print(attack.genres)
```

```python
# find the number of provoked attacks
provoked = session.query(Movies).filter_by(type='Provoked').count()
print(provoked)
```
